```
tf.compat.v1.reset_default_graph()
```

```
from tensorflow.python.keras.datasets import boston_housing

#load data
(features, actual_price), _ = boston_housing.load_data(test_split = 0)
```

```
features.shape
```

```
print(type(features))
```

```
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

features = sc.fit_transform(features)
```

```
features[0]
```

```
actual_price.shape[0]
```

**Building the Computational Graph**

```
tf.compat.v1.disable_eager_execution() 

x = tf.compat.v1.placeholder(shape = [None, 13], dtype = tf.float32, name = 'x_input')

y = tf.compat.v1.placeholder(shape = [None], dtype = tf.float32, name = 'y_input')
```

```
W = tf.Variable(tf.zeros(shape = [13, 1]), name = 'weights')

b = tf.Variable(tf.zeros(shape = [1]), name = 'bias')
```

```
y_hat = tf.add(tf.matmul(x, W), b, name = 'predicted_output')
```

```
loss = tf.reduce_mean(tf.square(y_hat - y), name = 'Loss')
```

```
train_op = tf.compat.v1.train.GradientDescentOptimizer(0.03).minimize(loss) # 0.03 is the learning rate
```

**Executing the Computational Graph**

```
sess = tf.compat.v1.Session()

sess.run(tf.compat.v1.global_variables_initializer())

training_epochs = 100
```

```
for epoch in range(training_epochs):
   _, training_loss = sess.run([train_op, loss], feed_dict = {x : features, y : actual_price})

   if epoch % 10 == 0:
     print("Epoch : {0}, cost : {1}".format(epoch, training_loss))
```
